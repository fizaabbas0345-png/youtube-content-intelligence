```python
 
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"F:/youtube-content-intelligence/data/raw/videos.csv")

df.head()
```

```python
print("Shape:", df.shape)

df.info()
```

```python
df.isnull().sum()
```

some missing values found in dataset

# Fill missing values

```python
df["like_count"] = df["like_count"].fillna(0)
```

fill the miss count value with 0

```python
df["comment_count"] = df["comment_count"].fillna(0)
```

fill the missing comment count value with zero

```python
median_subs = df["channel_subscribers"].median()

df["channel_subscribers"] = df["channel_subscribers"].fillna(median_subs)
```

fill missing values of channel_subsctibers with the median of the column

```python
df["tags"] = df["tags"].fillna("")
```

fill tag missing values with "" because tag type is object if we fill with nan it will fail the col

```python
df.isnull().sum()
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

duplicate video records were checked using the unique video_id.

```python
df["upload_date"] = pd.to_datetime(
    df["upload_date"],
    format="%Y%m%d"
)

df["upload_date"].head()
```

Upload dates were converted into datetime format

```python
df["year"] = df["upload_date"].dt.year

df["month"] = df["upload_date"].dt.month_name()

df["day"] = df["upload_date"].dt.day_name()
```

Additional temporal features (Year, Month, Day) were extracted to support publishing frequency analysis.

```python
# Duration in minutes

df["duration_minutes"] = (
    df["duration_seconds"] / 60
).round(2)
```

Video duration was converted from seconds to minutes to improve readability and simplify comparative analysis.

```python
df["engagement_rate"] = (
    (df["like_count"] + df["comment_count"])
    / df["views"]
) * 100

df["engagement_rate"] = df["engagement_rate"].round(2)
```

(Likes + Comments)/ Views×100


Engagement Rate was calculated to evaluate audience interaction relative to total views

```python
df["description_length"] = (
    df["description"]
    .fillna("")
    .str.len()
)
```

Description length was calculated to investigate whether detailed video descriptions are associated with higher audience engagement.

```python
df["title_length"] = (
    df["title"]
    .str.len()
)
```

Title length was calculated to analyze whether concise or descriptive titles influence video popularity.

```python
df["tag_count"] = (
    df["tags"]
    .fillna("")
    .apply(lambda x: len(x.split(",")) if x else 0)
)
```

Tag count was calculated to examine the relationship between metadata optimization and video performance.
